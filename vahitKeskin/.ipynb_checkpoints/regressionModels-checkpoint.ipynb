```python
import statsmodels.api as sm
```

```python
X   = df[["TV"]]
X[0:5]
```

```python
X = sm.add_constant(X)
X[0:5]
```

```python
y = df[["sales"]]
```

```python
lm = sm.OLS(y,X)
```

```python
model = lm.fit()
```

```python
model.summary()
```

```python
# farkl bir kullanm tr

import statsmodels.formula.api as smf
lm = smf.ols("sales ~ TV",df)
model = lm.fit()
model.summary()
```

```python
model.params
```

```python
model.summary().tables[1]
```

```python
model.conf_int()
```

```python
model.f_pvalue
```

```python
print("f_pvalue:", "%.3f" % model.f_pvalue)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["TV"]]
y = df[["sales"]]
reg = LinearRegression()
model = reg.fit(X,y)
model.intercept_
```

```python
model.coef_
```

```python
model.score(X,y) # r kare deeri
```
