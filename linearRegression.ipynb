```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pls
import seaborn as sns
```

```python
df = pd.read_excel("foms17_1yeardata.xlsx")
df.head()
```

```python
df['sonuc'] = np.random.randint(99, 102, size=len(df))

x = df[['CH1', 'CH2', 'CH3', 'CH4', 'CT', 'ST']]
y = df['sonuc']
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# 3. Veri Setini Bölmek
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# 4. Modeli Oluşturma ve Eğitme
model = LinearRegression()
model.fit(X_train, y_train)

# 5. Modeli Değerlendirme
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error (Hata Kareleri Ortalaması): {mse}")

# 6. Modeli Kullanma
# Örnek olarak, yeni bir veri noktasının sonucunu tahmin etmek için:
new_data = pd.DataFrame({'CH1': [97476], 'CH2': [45297], 'CH3': [35901],
                         'CH4': [33279], 'CT': [21.8], 'ST': [27.1]})
prediction = model.predict(new_data)
print(f"Sonuc Tahmini: {prediction}")
```

```python
# Katsayıları al
coefficients = model.coef_

# Sabiti al
intercept = model.intercept_

coefficients, intercept
```
