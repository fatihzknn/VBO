```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import make_blobs

df = pd.read_excel("TemizlenmisTerminalDatasi/foms11_motorin_300ppb.xlsx")

new_df = df[["CH1", "CH2", "CH3", "CH4"]]


new_df.head()
```

```python
# PCA modelinin oluşturulması ve uygulanması
pca = PCA(n_components=2)
data_transformed = pca.fit_transform(new_df)

# İndirgenmiş veriyi DataFrame'e dönüştürme
df_transformed = pd.DataFrame(data_transformed, columns=['PC1', 'PC2'])

```

```python
# Orjinal ve indirgenmiş veriyi görselleştirme
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.scatter(new_df['CH1'], new_df['CH2'])
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Orjinal Veri')

plt.subplot(1, 2, 2)
plt.scatter(df_transformed['PC1'], df_transformed['PC2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('İndirgenmiş Veri')

plt.tight_layout()
plt.show()
```
