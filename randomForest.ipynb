```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd

# Veri setini ykleme veya oluturma
cleaned_df = pd.read_excel("TemizlenmisTerminalDatasi/temizYllkVeri.xlsx")
X = cleaned_df[['CH1', 'CH2', 'CH3', 'CH4', 'CT', 'ST']]
y = cleaned_df['LM']
# X: Bamsz deikenler, y: Baml deiken

# Veriyi eitim ve test verisi olarak blmek
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Lineer Regresyon modelini oluturma ve eitme
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Tahminler yapma
y_pred = linear_model.predict(X_test)

# Grselletirmeler
plt.figure(figsize=(10, 6))

# Gerek deerler
plt.scatter(y_test, y_pred, color='blue', alpha=0.7, label='Gerek vs Tahmin')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', linewidth=2, label='Doru')

plt.xlabel('Gerek Deerler')
plt.ylabel('Tahmin Edilen Deerler')
plt.title('Gerek ve Tahmin Edilen Deerler Karlatrmas')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

# Veri setini ykleme veya oluturma
cleaned_df = pd.read_excel("TemizlenmisTerminalDatasi/temizYllkVeri.xlsx")
X = cleaned_df[['CH1', 'CH2', 'CH3', 'CH4', 'CT', 'ST']]
y = cleaned_df['LM']
# X: Bamsz deikenler, y: Baml deiken

# Veriyi eitim ve test verisi olarak blmek
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Rastgele Orman Regresyon modelini oluturma ve eitme
rf_model = RandomForestRegressor(n_estimators=300, random_state=0)  # rnek olarak 100 aa kullanld
rf_model.fit(X_train, y_train)

# Tahminler yapma
y_pred = rf_model.predict(X_test)

# Grselletirmeler
plt.figure(figsize=(10, 6))

# Gerek deerler
plt.scatter(y_test, y_pred, color='blue', alpha=0.7, label='Gerek vs Tahmin')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', linewidth=2, label='Doru')

plt.xlabel('Gerek Deerler')
plt.ylabel('Tahmin Edilen Deerler')
plt.title('Gerek ve Tahmin Edilen Deerler Karlatrmas')
plt.legend()
plt.show()
```

```python
y_pred
```

```python
residuals = y_test - y_pred

plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')
plt.xlabel('Tahmin Edilen Deerler')
plt.ylabel('Hatalar (Gerek - Tahmin)')
plt.title('Hata Dalm')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.hist(y_test, bins=20, color='blue', alpha=0.7, label='Gerek Deerler')
plt.hist(y_pred, bins=20, color='orange', alpha=0.7, label='Tahmin Edilen Deerler')
plt.xlabel('Deerler')
plt.ylabel('Frekans')
plt.title('Gerek ve Tahmin Edilen Deer Dalmlar')
plt.legend()
plt.show()
```
