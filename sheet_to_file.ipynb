```python
import openpyxl
import os

# Kaydetmek istediğiniz Excel dosyasının adını belirtin
original_excel_file = 'yeni_test_data/test_data.xlsx'

# Excel dosyasını yükle
workbook = openpyxl.load_workbook(original_excel_file)

# Excel sayfalarını al
sheet_names = workbook.sheetnames

# Her bir sayfa için ayrı Excel dosyası oluştur
for sheet_name in sheet_names:
    # Yeni bir çalışma kitabı oluştur
    new_workbook = openpyxl.Workbook()
    new_sheet = new_workbook.active
    
    # Verileri kopyala
    original_sheet = workbook[sheet_name]
    for row in original_sheet.iter_rows():
        new_sheet.append([cell.value for cell in row])
    
    # Yeni Excel dosyasını kaydet
    new_excel_file = f'yeni_test_data/{sheet_name}.xlsx'
    new_workbook.save(new_excel_file)
    
    print(f'{sheet_name} sayfası {new_excel_file} dosyasına kaydedildi.')

# Orijinal Excel dosyasını kapat
workbook.close()

print('İşlem tamamlandı.')
```
