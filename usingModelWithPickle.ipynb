```python
import pickle
import pandas as pd
import numpy as np
import os

# Modeli geri yükleme
with open('model.pkl', 'rb') as model_file:
    loaded_model = pickle.load(model_file)
    
folder_path = "yeni_test_data/"
output_folder = "yeni_test_data_guncel/"

for filename in os.listdir(folder_path):
    if filename.endswith(".xlsx"):
        print(filename)
        file_path = os.path.join(folder_path, filename)
        
        # Dosya adındaki boşlukları ve çizgileri düzgün bir şekilde işlemleyin
        clean_filename = filename.replace(" ", "").replace("-", "")
        
        df = pd.read_excel(file_path)
        
        
        df.columns = ["time", "CH1","CH2","CH3","CH4","ft","st","label"]
       
        df.replace("\n", np.nan, inplace=True)

        # NaN değerleri temizleme
        df.dropna(inplace=True)
        
        df_selected = df[["CH1","CH2","CH3","CH4"]]
        
        
        # Tahminleri yapma
        predictions = loaded_model.predict(df_selected)
        
        # Tahminleri DataFrame'e ekleyin
        df["Prediction"] = predictions
        
        # Yeni dosya adı ve yolunu oluşturun
        output_file_path = os.path.join(output_folder, f"{clean_filename}_guncel.xlsx")
        
        # Veriyi Excel dosyasına kaydetme
        df.to_excel(output_file_path, index=False)
```
